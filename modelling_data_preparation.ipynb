For the modelling dataset, the following transformations were implemented:

- the dataset was pivoted
- the Year was converted to datetime format (yearly) was set as 'index'
- the column header was tagged 'Area Code (ISO3)'
- the average emission column was added for the respective years

```python
# import libraries
import pandas as pd
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Set display options to show full dataset
pd.set_option('display.max_rows', None)   # Show all rows
pd.set_option('display.max_columns', None) # Show all columns
pd.set_option('display.width', None)      # Do not truncate output
```

```python
# load the cleaned dataset
data = pd.read_csv("dataset/Cattle_CH4_dataset_cleaned_2000_2021.csv")
data.head()
```

```python
# Pivot the DataFrame
df = data.pivot(index='Year', columns='Area Code (ISO3)', values='Value')

# Remove the column name - 'Area Code (ISO3)'
df.columns.name = None

# Convert the index to datetime format and rename the index
df.index = pd.to_datetime(df.index, format='%Y')
df.index.name = 'date'

# Add "avg_emission" column for all countries
df["avg_emission"]=df.mean(axis=1)

# Display the first 5 rows of the dataset
df.head()
```

```python
# save file to csv
df.to_csv("dataset/modelling_data_2000_2021.csv")
```
